```python
# =============================================================================
# TUGAS PRAKTIKUM BAB 4
# Deteksi Clickbait pada Judul Berita Bahasa Indonesia
# Menggunakan Multi-Input Neural Network (LSTM + MLP)
# =============================================================================
# Nama   : Nurul Annisa
# Nim    : 230705176
# =============================================================================

# ─────────────────────────────────────────────────────────────────────────────
# SECTION 1: Import Modul dan Dataset
# ─────────────────────────────────────────────────────────────────────────────
import re
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.utils import class_weight
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.models import Model
from keras.layers import Input, Dense, LSTM, Dropout, Embedding, concatenate

import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

# ─────────────────────────────────────────────────────────────────────────────
# SECTION 2: Load Dataset
# ─────────────────────────────────────────────────────────────────────────────
# Ganti path berikut sesuai lokasi file di komputermu
df = pd.read_csv('all_agree.csv')

print("=== 5 Data Pertama ===")
print(df.head())

print("\n=== Shape Dataset ===")
print(df.shape)

print("\n=== Distribusi Label ===")
print(df['label'].value_counts())

# ─────────────────────────────────────────────────────────────────────────────
# SECTION 3: Feature Generation
# Membuat fitur-fitur baru dari kolom title
# ─────────────────────────────────────────────────────────────────────────────

# Fungsi 1: Menghitung tanda baca (punctuation)
def find_punc(x, punc):
    count = 0
    for char in x:
        if char == punc:
            count += 1
    return count

# Fungsi 2: Menghitung huruf kapital
def find_uppercase(x):
    count = 0
    for char in x:
        if char.isupper():
            count += 1
    return count

# Fungsi 3: Menghitung huruf kecil
def find_lowercase(x):
    count = 0
    for char in x:
        if char.islower():
            count += 1
    return count

# Fungsi 4: Menghitung angka
def find_number(x):
    count = 0
    for char in x:
        if char in [str(i) for i in range(10)]:
            count += 1
    return count

# Terapkan semua fungsi ke dataframe
df['punc_count']      = df['title'].apply(lambda x: find_punc(x, ','))
df['fullstop_count']  = df['title'].apply(lambda x: find_punc(x, '.'))
df['question_count']  = df['title'].apply(lambda x: find_punc(x, '?'))
df['exclaim_count']   = df['title'].apply(lambda x: find_punc(x, '!'))
df['uppercase_count'] = df['title'].apply(lambda x: find_uppercase(x))
df['lowercase_count'] = df['title'].apply(lambda x: find_lowercase(x))
df['number_count']    = df['title'].apply(lambda x: find_number(x))

print("\n=== Shape setelah Feature Generation ===")
print(df.shape)   # Harusnya (8613, 10)
print(df.head())

# ─────────────────────────────────────────────────────────────────────────────
# SECTION 4: Membersihkan Data di Kolom title
# ─────────────────────────────────────────────────────────────────────────────

def clean(text):
    # Lowercase semua huruf
    text = text.lower()
    # Hapus tanda baca
    text = re.sub(r'[!?,.:;\'"()-]', ' ', text)
    # Tokenisasi menggunakan NLTK
    word_tokens = word_tokenize(text)
    # Gabung kembali menjadi string
    text = ' '.join(word_tokens)
    return text

df['cleaned_title'] = df['title'].apply(clean)

print("\n=== Contoh Hasil Cleaning ===")
print(df[['title', 'cleaned_title']].head())
print("\n=== Shape setelah Cleaning ===")
print(df.shape)   # Harusnya (8613, 11)

# ─────────────────────────────────────────────────────────────────────────────
# SECTION 5: Train-Test Split
# ─────────────────────────────────────────────────────────────────────────────

# X = semua fitur mulai dari kolom ke-4 (index 3) ke kanan
# y = kolom label_score (0 = non-clickbait, 1 = clickbait)
X = df.iloc[:, 3:]
y = df['label_score']

print("\n=== Kolom X ===")
print(X.columns.tolist())

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=77
)

print("\n=== Shape Train & Test ===")
print("X_train:", X_train.shape)   # (6890, 8)
print("X_test :", X_test.shape)    # (1723, 8)

# ─────────────────────────────────────────────────────────────────────────────
# SECTION 6: Preprocessing Lanjutan untuk LSTM dan MLP
# ─────────────────────────────────────────────────────────────────────────────

# --- 6a. Preprocessing Teks untuk Input LSTM ---

X_train_title = X_train['cleaned_title']
X_test_title  = X_test['cleaned_title']

# Tokenisasi: ubah kata menjadi angka integer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train_title)

X_train_title = tokenizer.texts_to_sequences(X_train_title)
X_test_title  = tokenizer.texts_to_sequences(X_test_title)

vocab_size = len(tokenizer.word_index)
print("\n=== Jumlah Kata Unik (vocab_size) ===")
print(vocab_size)   # Sekitar 12265

# Cari panjang kalimat terpanjang untuk padding
lengths = []
for sentence in X_train_title:
    lengths.append(len(sentence))

print("\n=== Statistik Panjang Kalimat ===")
print("Terpanjang :", max(lengths))
print("Terpendek  :", min(lengths))
print("Rata-rata  :", np.mean(lengths))

maxlen = max(lengths)   # = 19

# Padding: samakan panjang semua kalimat menjadi maxlen
X_train_title = pad_sequences(X_train_title, maxlen=maxlen)
X_test_title  = pad_sequences(X_test_title,  maxlen=maxlen)

print("\n=== Shape setelah Padding ===")
print("X_train_title:", X_train_title.shape)   # (6890, 19)
print("X_test_title :", X_test_title.shape)    # (1723, 19)

# --- 6b. Preprocessing Fitur Numerik untuk Input MLP ---

# Pisahkan fitur numerik (buang kolom teks)
X_train_punc = X_train.drop(['cleaned_title'], axis=1)
X_test_punc  = X_test.drop(['cleaned_title'], axis=1)

print("\n=== Shape Fitur Numerik ===")
print("X_train_punc:", X_train_punc.shape)   # (6890, 7)
print("X_test_punc :", X_test_punc.shape)    # (1723, 7)

# ─────────────────────────────────────────────────────────────────────────────
# SECTION 7: Pembobotan Class (Class Weighting)
# Mengatasi data imbalanced (non-clickbait lebih banyak dari clickbait)
# ─────────────────────────────────────────────────────────────────────────────

class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train),
    y=y_train.values
)
class_weight_dict = dict(enumerate(class_weights))

print("\n=== Class Weights ===")
print(class_weight_dict)
# Contoh: {0: 0.81..., 1: 1.30...}
# Artinya class clickbait (1) diberi bobot lebih besar

# ─────────────────────────────────────────────────────────────────────────────
# SECTION 8: Membangun Multi-Input Neural Network
# Arsitektur: LSTM (input teks) + MLP (input fitur numerik) → Concatenate
# ─────────────────────────────────────────────────────────────────────────────

tf.random.set_seed(99)

num_features = X_train_punc.shape[1]   # = 7

# --- Cabang 1: LSTM untuk input teks ---
lstm_input = Input(shape=(maxlen,))
stm        = Embedding(vocab_size + 1, 32, input_length=maxlen)(lstm_input)
stm        = LSTM(32, return_sequences=False)(stm)
stm        = Dropout(rate=0.5)(stm)
lstm_model = Model(inputs=lstm_input, outputs=stm)

# --- Cabang 2: MLP untuk input fitur numerik ---
dense_input  = Input(shape=(num_features,))
dense        = Dense(3, activation='sigmoid')(dense_input)
dense        = Dropout(rate=0.5)(dense)
dense_model1 = Model(inputs=dense_input, outputs=dense)

# --- Gabungkan kedua cabang ---
concat  = concatenate([lstm_model.output, dense_model1.output])
dense2  = Dense(6, activation='sigmoid')(concat)
dense3  = Dense(3, activation='sigmoid')(dense2)
output  = Dense(1, activation='sigmoid')(dense3)   # Output: 0 (non-clickbait) atau 1 (clickbait)

model = Model(
    inputs=[lstm_model.input, dense_model1.input],
    outputs=output
)

# --- Compile ---
model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['acc']
)

print("\n=== Ringkasan Arsitektur Model ===")
model.summary()

# ─────────────────────────────────────────────────────────────────────────────
# SECTION 9: Training dan Evaluasi Model
# ─────────────────────────────────────────────────────────────────────────────

print("\n=== Mulai Training ===")
history = model.fit(
    [X_train_title, X_train_punc],
    y_train,
    batch_size=64,
    epochs=30,
    class_weight=class_weight_dict,
    validation_data=(
        [X_test_title, X_test_punc],
        y_test
    )
)

# --- Plot Akurasi ---
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['acc'],     label='Train Accuracy')
plt.plot(history.history['val_acc'], label='Val Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# --- Plot Loss ---
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'],     label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig('training_curves.png', dpi=150)
plt.show()
print("Grafik disimpan sebagai 'training_curves.png'")

# --- Evaluasi Akhir ---
print("\n=== Evaluasi Model pada Data Test ===")
test_loss, test_acc = model.evaluate(
    [X_test_title, X_test_punc], y_test, verbose=0
)
print(f"Test Loss    : {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.4f} ({test_acc*100:.2f}%)")
print("Target akurasi referensi: ~88.28%")

# --- Confusion Matrix ---
y_pred_prob = model.predict([X_test_title, X_test_punc])
y_pred      = (y_pred_prob > 0.5).astype(int).flatten()

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(
    cm, annot=True, fmt='d', cmap='Blues',
    xticklabels=['Non-Clickbait', 'Clickbait'],
    yticklabels=['Non-Clickbait', 'Clickbait']
)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=150)
plt.show()
print("Confusion matrix disimpan sebagai 'confusion_matrix.png'")

print("\n=== SELESAI ===")
```
